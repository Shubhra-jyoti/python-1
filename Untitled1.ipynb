```python
f=open("mbox-short.txt")
x=f.readlines()
fx=[]
for i in x:
    if(i.startswith("From:")):
        i=i.split()
        fx.append(i[1])
for i in fx:
    print(i)
print(len(fx),"adderesses")    
f.close()
```

```python
f1=open("mbox-short.txt")
x=f1.readlines()
user={}
for i in x:
    if(i.startswith("From:")):
            s=i.split()[1]
            user[s]=user.get(s,0)+1
max1=max(user.values()) 
for i in user:
    if(user[i]==max1):
        print(i)
f1.close()
```

```python
f=open("logfile.txt")
x=f.readlines()
for i in x:
    s=i.split(" ")
    st=s[1]
    et=s[2]
    st=st.split(":")[0]
    et=et.split(":")[0]
    st=int(st)
    et=int(et)
    duration=et-st
    if(duration>=1):
        print(i)
        
```
