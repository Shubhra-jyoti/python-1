```python
f=open("3.txt","r+")
# f=open("3.txt","r+") while using r+mode first read then write or else ovveerwruite will start

f.write("vha")
print(f.read())

f.close()
```

```python
f=open("3.txt","w+")
f.write("Hello")
print(f.tell())
f.seek(0)#seek takes the pointer to a specific location
print(f.tell())
print(f.read())
f.seek(0,2)
f.write("welcome")
f.close()
```

```python
with open("2.txt") as f:
    print(f.name)
    print(f.mode)
    print(f.closed)
    print(f.readable())
    print(f.writable())
      
```
