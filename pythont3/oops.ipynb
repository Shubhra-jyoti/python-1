```python
class Atm:
    def __init__(self):
        self.balance=0
        self.pin=""
        self.menu()
    def createpin(self,pin,balance):
        self.pin=pin
        self.balance=balance
    def updatepin(self):
        pin=input("Enter current pin")
        if(self.pin==pin):
            pin=input("Enter new pin")
            self.pin=pin
            print("pin updated")
        else:
            
            print("invalid pin")
    def withdraw(self,amount):
        if(self.balance<amount or self.balance-amount<1000):
            print("insufficient balance")
        else:
            self.balance-=amount
            print("withdraw successfull")
    def menu(self):
        while True:
            print("1.Create pin and enter balance")
            print("2.update pin")
            print("3.withdraw")
            print("4.Exit")
            ch=int(input("Enter choice"))
            if ch==1:
                pin=input("Enter pin")
                balance=float(input("Enter balance"))
                if(balance<0):
                    print("invalid")
                else:
                    self.createpin(pin,balance)
            elif(ch==2):
                self.updatepin()
            elif ch==3:
                amt=float(input("Enter amount"))
                if(amt<0):
                    print("Invalid amount")
                else:
                    self.withdraw(amt)
            elif ch==4:
                print("Thank you server down")
                break
            else:
                print("invalid choice")
                
                    
sbi=Atm()            
            

            
            
        
```
