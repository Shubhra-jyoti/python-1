```python
def storecustomer():
    f=open("customer.txt","a")
    name=input("Enter Name")
    age=input("Enter age")
    f.writelines([name+"\t",age+"\n"])
    f.close()
storecustomer()    
```

```python
f=open("friends.txt","a+")
f.seek(0)
x=f.readlines()
print(len(x),"number of lines")
charct=0
f.seek(0)
for i in f.read():
    charct+=1
print(charct,"number of words")
f.seek(0)
words=len((f.read()).split())
print(words,"number of words")

f.close()
```

```python
def display_oddlines():
    f=open("friends.txt")
    x=f.readlines()
    x=x[0::2]
    for i in x:
        print(i)
print(x)        
display_oddlines()        
```
