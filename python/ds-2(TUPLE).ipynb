```python
values=eval(input("Enter tuple"))
even=()
odd=()
for i in values:
    if(i%2==0):
        even=even+(i,)
    else:
        odd=odd+(i,)
print(even) 
print(odd)
print("Mx_even=",max(even))
print("Mx_odd=",max(odd))
print("Min_even=",min(even))
print("Min_odd=",min(odd))

```
