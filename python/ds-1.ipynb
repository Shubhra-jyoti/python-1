```python
s=input("Enter string")
s=s.title()[:-1]+s[-1].upper()
counter=0
for i in s:
    if(i.isspace()):
        counter-=1
        break
    counter+=1 
s=s[:counter]+s[counter].upper()+s[counter+1:]    
print(s)
    
```
